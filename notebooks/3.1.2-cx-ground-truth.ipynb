## Ellipsoid

```python
def generate_ellipsoid_cloud(a, b, c, num_points = 5000):
    """Generate a random point on an ellipsoid defined by a,b,c"""
    
    theta = np.random.uniform(0, 2*np.pi, num_points)
    v = np.random.rand(num_points)
    phi = np.arccos(2.0 * v - 1.0)
    sinTheta = np.sin(theta);
    cosTheta = np.cos(theta);
    sinPhi = np.sin(phi);
    cosPhi = np.cos(phi);
    rx = a * sinPhi * cosTheta;
    ry = b * sinPhi * sinTheta;
    rz = c * cosPhi;
    return np.column_stack((rx, ry, rz))

def ellipsoid_ground_truth(point_cloud, a, b, c):
    #point_cloud: N by 3 array
    cur = []
    for p in point_cloud:
        k = 1 / (a **2 * b**2 * c**2 * (p[0]**2 / a**4 + p[1]**2 / b**4 + p[2]**2 / c**4) **2)
        cur.append(k)
    return cur
        
```

```python
ellipsoid = generate_ellipsoid_cloud(0.9, 1.25, 0.9)

# Visualize the point cloud
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(ellipsoid[:, 0], ellipsoid[:, 1], ellipsoid[:, 2], s=5)
ax.set_aspect('equal')
ax.set_title("Generated Sphere Point Cloud")
ax.view_init(0, 0)
plt.axis('off')
plt.show()
```

```python
cc = ellipsoid_ground_truth(ellipsoid, 0.9, 1.25, 0.9)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(ellipsoid[:, 0], ellipsoid[:, 1], ellipsoid[:, 2], s=5, c = cc)
ax.set_aspect('equal')
ax.set_title("Generated Sphere Point Cloud")
ax.view_init(0, 0)

plt.show()
```

## Torus 

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def generate_torus_point_cloud(num_points = 5000, R = 3, r = 1):
    # Generate random angles for theta and phi
    theta = np.random.uniform(0, 2*np.pi, num_points)
    phi = np.random.uniform(0, 2*np.pi, num_points)

    # Compute the torus points
    x = (R + r * np.cos(phi)) * np.cos(theta)
    y = (R + r * np.cos(phi)) * np.sin(theta)
    z = r * np.sin(phi)

    K = np.cos(phi)/(r * (R + r * np.cos(phi))) 

    return np.column_stack((x, y, z)), K

# Create torus parameters
R = 1 # Major radius
r = 0.375  # Minor radius
num_samples = 5000

# Generate a sphere point cloud with 1000 points and radius 1
torus_cloud, torus_K = generate_torus_point_cloud(num_points = num_samples, R = R, r = r)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(torus_cloud[:, 0], torus_cloud[:, 1], torus_cloud[:, 2], s=5, c = torus_K)
# Add colorbar
cbar = fig.colorbar(sc, ax=ax)
cbar.set_label('K')
ax.set_aspect('equal')
ax.set_title("Generated Torus Point Cloud")
ax.view_init(elev = 3)
```

```python
def generate_hyperboloid_cloud(a, b, c, num_points=5000):
    
    u = np.random.uniform(-1.5,1.5,num_points)
    v = np.random.uniform(0, 2*np.pi, num_points)

    x = a*np.sqrt(1 + np.square(u)) * np.cos(v)
    y = b*np.sqrt(1 + np.square(u)) * np.sin(v)
    z = c*u

    K = - c**2 / np.square(c**2 + (a**2 + c**2) * np.square(u))
    return np.column_stack((x, y, z)), K

# Create hyperboloid parameters
a = 1
b = 1 
c = 1
num_samples = 5000

# Generate a hyperboloid point cloud with 1000 points and radius 1
hyperboloid_cloud, hyperboloid_K = generate_hyperboloid_cloud(a, b, c, num_points = num_samples)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(hyperboloid_cloud[:, 0], hyperboloid_cloud[:, 1], hyperboloid_cloud[:, 2], s=5, c = hyperboloid_K)
# Add colorbar
cbar = fig.colorbar(sc, ax=ax)
cbar.set_label('K')
ax.set_aspect('equal')
ax.set_title("Generated Hyperboloid Point Cloud")
ax.view_init(elev = 3)
```
