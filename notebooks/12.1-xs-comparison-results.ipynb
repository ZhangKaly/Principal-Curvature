```python
import pandas as pd 
import wandb
import numpy as np

api = wandb.Api()
sweep = api.sweep("xingzhis/sheaf_curvature/2nrr8fvi")

sweep_runs_data = []
for run in sweep.runs:
    # Fetch run summary (metrics)
    run_data = {"id": run.id, "name": run.name}
    for key, value in run.summary.items():
        run_data["metric_" + key] = value

    # Fetch run configuration (parameters)
    for key, value in run.config.items():
        run_data["param_" + key] = value

    sweep_runs_data.append(run_data)

df = pd.DataFrame(sweep_runs_data)
```

```python
def create_pivot_table(df, path_name, metric_name='metric_coef'):
    filtered_df = df[df['param_path.name'] == path_name]
    pivot_table = filtered_df.pivot(index='param_path.noise', columns='param_method.name', values=metric_name)
    return pivot_table

# Creating three tables for 'ellipsoid', 'hyperboloid', and 'torus'
ellipsoid_table = create_pivot_table(df, 'ellipsoid')
hyperboloid_table = create_pivot_table(df, 'hyperboloid')
torus_table = create_pivot_table(df, 'torus')
```
